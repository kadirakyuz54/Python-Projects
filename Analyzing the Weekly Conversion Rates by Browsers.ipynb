# Analyzing the Weekly Conversion Rates by Browsers

#### In this project I will analyze the conversion rates by browsers and countries. We have the number of visits and conversions for 20 weeks for 4 different browsers in 10 different countries.

Conversion rate is the percentage of visitors to your website that complete a desired goal out of the total number of visitors. A high conversion rate is indicative of successful marketing and web design: It means people want what you're offering, and they're easily able to get it! The results indicate that over time the browsers have been increasingly doing a better job for marketing.

```python
import pandas as pd
conversion = pd.read_csv('conversion_rates.csv')
conversion['visits'] = conversion['visits'].astype('float')
conversion['conversions'] = conversion['conversions'].astype('float')
conversion['week'] = conversion['week'].astype('int')
conversion.head()
```

```python
weekly_conv_counts = conversion.groupby('week').sum()
weekly_conv_counts
```

```python
weekly_conv_counts['weekly_conv_rates'] = weekly_conv_counts['conversions']/weekly_counts['visits']
weekly_conv_counts
```

The distribution of the weekly conversion rates over time is as follows. Overall, there is an increasing trend.

```python
avg_weekly_conv_r = weekly_conv_counts['weekly_conv_rates'].mean()
avg_weekly_conv_r
```

The average weekly conversion rate across all countries and browsers over time is 38.22%

```python
weekly_con_by_browser = conversion.groupby(['week', 'browser']).sum()
weekly_con_by_browser
```

```python
weekly_con_by_browser['weekly_rates_by_browser'] = weekly_con_by_browser['conversions']/weekly_con_by_browser['visits']
weekly_con_by_browser['weekly_rates_by_browser']
```

```python
weekly_con_by_browser['weekly_rates_by_browser'].groupby('browser').plot(legend='browser')
```

```python
weekly_con_by_country = conversion.groupby(['country', 'week']).sum()
weekly_con_by_country
```

```python
weekly_con_by_country['weekly_rates_by_country'] = weekly_con_by_country['conversions']/weekly_con_by_country['visits']

weekly_con_by_country['weekly_rates_by_country']

weekly_con_by_country['weekly_rates_by_country'].groupby('country').mean().sort_values(ascending=False) 
```

```python
weekly_con_by_country['weekly_rates_by_country'].groupby('country').plot(legend='country')

plt.title('Conversion Rates by Countries over Time')
plt.xlabel('Week')
plt.ylabel('Rates')
plt.show()
```

```python
import seaborn as sns
rates_by_country = weekly_con_by_country.groupby('country').mean().sort_values('weekly_rates_by_country', ascending=False)
rates_by_country
```

```python
sns.catplot(x= rates_by_country.index, y= rates_by_country['weekly_rates_by_country'],data=rates_by_country, kind='bar')

plt.xlabel('Country')
plt.ylabel('Average Conversion Rates')
plt.title("Average Conversion Rates by Country")
```

Average weekly conversion rates by countries are shown in the figure. Accordingly, the highest conversion rate belongs to AE with 81.31% and it is followed by AC with 74.14%.

```python
weekly_con_by_browser = conversion.groupby(['week', 'browser']).mean()
weekly_con_by_browser['weekly_rates_by_browser'] = weekly_con_by_browser['conversions']/weekly_con_by_browser['visits']
weekly_con_by_browser

rates_by_browser = weekly_con_by_browser.groupby('browser').mean().sort_values('weekly_rates_by_browser', ascending=False)
rates_by_browser
```
